```sql
CREATE OR REPLACE VIEW gold.petrol_prices.latest_prices_metric_view
WITH METRICS
LANGUAGE YAML
AS $$
  version: 1.1
  source: gold.petrol_prices.latest_prices
  joins:
    - name: postcodes_dim
      source: gold.petrol_prices.postcodes
      "on": source.postcode = postcodes_dim.postcode

  dimensions:
    - name: name
      expr: source.name
      display_name: Name
    - name: forecourt_id
      expr: source.forecourt_id
      display_name: Forecourt Id
    - name: trading_name
      expr: source.trading_name
      display_name: Trading Name
    - name: brand_name
      expr: source.brand_name
      display_name: Brand Name
    - name: forecourt_type
      expr: source.forecourt_type
      display_name: Forecourt Type
    - name: postcode
      expr: source.postcode
      display_name: Postcode
    - name: super_unleaded
      expr: source.super_unleaded
      display_name: Super Unleaded
    - name: unleaded
      expr: source.unleaded
      display_name: Unleaded
    - name: premium_diesel
      expr: source.premium_diesel
      display_name: Premium Diesel
    - name: diesel
      expr: source.diesel
      display_name: Diesel
    - name: biodiesel
      expr: source.biodiesel
      display_name: Biodiesel
    - name: hydrogen
      expr: source.hydrogen
      display_name: Hydrogen
    - name: geographic_postcode
      expr: postcodes_dim.geographic_postcode
    - name: outward_postcode
      expr: postcodes_dim.outward_postcode
    - name: latitude
      expr: postcodes_dim.latitude
    - name: longitude
      expr: postcodes_dim.longitude

  measures:
    - name: count
      expr: COUNT(*)
      display_name: Count
    - name: median_unleaded_price
      expr: MEDIAN(source.unleaded)
      display_name: Median Unleaded Price
$$
```
