```sql
CREATE OR REPLACE VIEW gold.petrol_prices.latest_prices_metric_view
WITH METRICS
LANGUAGE YAML
AS $$
  version: 1.1

  source: gold.petrol_prices.latest_prices

  joins:
    - name: postcodes_dim
      source: gold.petrol_prices.postcodes
      "on": source.postcode = postcodes_dim.postcode
    - name: fuel_types_dim
      source: gold.petrol_prices.fuel_types
      "on": source.fuel_type_code = fuel_types_dim.fuel_type_code

  dimensions:
    - name: name
      expr: source.name
      display_name: Name
    - name: forecourt_id
      expr: source.forecourt_id
      display_name: Forecourt Id
    - name: trading_name
      expr: source.trading_name
      display_name: Trading Name
    - name: brand_name
      expr: source.brand_name
      display_name: Brand Name
    - name: forecourt_type
      expr: source.forecourt_type
      display_name: Forecourt Type
    - name: postcode
      expr: source.postcode
      display_name: Postcode
    - name: fuel_type_description
      expr: fuel_types_dim.fuel_type_description
      display_name: Fuel Type
    - name: price
      expr: source.price
      display_name: Price (pence per litre)
    - name: geographic_postcode
      expr: postcodes_dim.geographic_postcode
    - name: outward_postcode
      expr: postcodes_dim.outward_postcode
    - name: latitude
      expr: postcodes_dim.latitude
    - name: longitude
      expr: postcodes_dim.longitude

  measures:
    - name: count
      expr: COUNT(*)
      display_name: Count
    - name: median_price
      expr: MEDIAN(source.price)
      display_name: Median Price
$$
```
